**exercicio 5.3 — **kwargs**
Imos usar `kwargs`para:
- a función `demo_kwargs(**kwargs)` que devolva directamente `kwargs`. Proba con `nome='Iria', idade=17`.
- a función `saudar(**kwargs)` que recibe un número variable de parámetros. Entre eles:
  - `nome`: por defecto 'descoñecido'.
  - `apelidos`: por defecto ''.
  - `idioma`: pode ser `es`|`gl`|`en`, por defecto `gl`.
  
  O programa debe saudar, mencionando nome e apelidos, no idioma indicado. 

```python
# código do alumnado
def demo_kwargs(**kwargs):
    return kwargs
demo_kwargs(nome='Iria', idade=17)

def saudar(**kwargs):
    nome = kwargs.get('nome')
    apelidos = kwargs.get('apelidos','')
    idioma = kwargs.get('idioma','gl')

    if (idioma == 'gl'):
        return f"Ola, {nome} {apelidos}!"
    else:
        if (idioma == 'es'):
            return f"Hola, {nome} {apelidos}!"
        else:
            if (idioma == 'en'):
                return f"Hello, {nome} {apelidos}"
            else:
                return "Idioma non dispoñible"

saudar(nome='Adri',apelidos='Blanco', idioma='pt')        
```

**exercicio 5.4 — media segura**

Escribe `media(lista)` que devolva a media dunha lista de números, devolvendo `None` se a lista está baleira.

```python
# código do alumnado
# from typing import Optional, Sequence
# def media(lista: Sequence[float]) -> Optional[float]:
#     ...
# media([1,2,3]), media([])

def media (lista):
    if (len(lista)== 0):
        return None
    total = 0
    for i in lista:
        total+=i
    return total/len(lista)

media([1,2,3,4]), media([])
```

**exercicio 6.1 — clase Punto**

Crea unha clase `Punto` con `x` e `y`. Engade:
- método `desprazar(dx, dy)`
- método `modulo()` que devolva a distancia á orixe. Módulo = sqrt(x² + y²)
- `__repr__` amigable
Proba con `Punto(3,4)` e despraza (-1,2).


```python
class Punto:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def desprazar(self, dx, dy):
        self.x+=dx
        self.y+=dy
    def modulo(self):
        return (self.x**2 + self.y**2)**0.5
    
    def __repr__(self):
        return f"Punto({self.x},{self.y})"
#         ...
#     def modulo(self):
#         ...
#     def __repr__(self):
#         ...
p = Punto(3,4)
p.modulo()
p.desprazar(-1,2)
p
```

**exercicio 6.2 — herdar Punto3D**

Crea `Punto3D(Punto)` engadindo `z` e redefinindo `modulo()` para 3D. Proba con `(1,2,2)`. Módulo = sqrt(x² + y² + z²)

```python
# código do alumnado
# class Punto3D(Punto):
#     def __init__(self, x,y,z):
#         ...
#     def modulo(self):
#         ...
# p3 = Punto3D(1,2,2)
# p3, p3.modulo()

class Punto3D(Punto):
    def __init__(self,x,y,z):
        super().__init__(x,y)
        self.z = z
    
    def modulo(self):
        return (self.x**2 + self.y**2 + self.y**2)**0.5
    
p3d = Punto3D(1,2,2)

p3d.modulo()
        

```

**exercicio 6.3 — clase Alumno**

Crea `Alumno` con `nome` e `notas` (lista). Engade:
- `engadir(nota)`
- `media()` (ou `None` se baleira)
- `__repr__`
Proba con `Iria` e notas `[7,8,9]`.

```python
class Alumno:
    def __init__(self, nome, notas=None):
        self.nome = nome
        if (notas == None):
            self.notas=[]
        else:
            self.notas = notas
    
    def engadir(self, nota):
        self.notas.append(nota)

    def media (self):
        acc = 0
        for nota in self.notas:
            acc += nota
        return acc/len(self.notas)
    
    def __repr__(self):
        return f"Notas de {self.nome} : {self.notas}. Media: {self.media():.2f}"
    
a = Alumno('Iria',[6,3])
a.engadir(8)
print(a)
```
