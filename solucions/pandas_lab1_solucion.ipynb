## 3. Selección de datos con `iloc` e `loc`

### Exercicio 7: Selección con `iloc`
1) Primeira fila. 
2) Valor da terceira fila e segunda columna. 
3) Dúas primeiras filas completas.

```python
# TODO: usa iloc nas tres peticións

# 7.1
display(df.iloc[0,:])

# 7.2
df.iloc[2,1]

# 7.3
df.iloc[:1]
```

### Exercicio 8: Selección con `loc`
1) Fila coa etiqueta de índice 2. 
2) Todas as filas con columnas `nome` e `nota`. 
3) Filas 1 a 3 incluídas.

```python
# TODO: usa loc para as tres peticións
# 8.1
display(df.loc[2])

# 8.2
display(df.loc[:,['nome','nota']])

# 8.3
display(df.loc[1:3])
```

## 4. Filtrado de datos

### Exercicio 9: Filtros sinxelos
1) Alumnos con `nota > 8`. 
2) Alumnos con `idade == 20`.

```python
# TODO: crea dous DataFrames filtrando con condicións sinxelas

# 9.1
df[df['nota'] > 8]

# 9.2
df[df['idade'] == 20]
```

### Exercicio 10: Filtros combinados
1) `idade > 20` e `nota >= 8`. 
2) `idade` en [20, 22].

```python
# TODO: combina condicións con & e usa .isin([...])

# 10.1
df[(df['idade'] > 20) & (df['nota'] >= 8)]

# 10.2
df[df['idade'].isin([20,22])]
```
